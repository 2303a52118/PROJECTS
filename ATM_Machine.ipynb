```python
import os

class Account:
    def __init__(self, acc_no, name, pin, balance=0):
        self.acc_no = acc_no
        self.name = name
        self.pin = pin
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Amount deposited successfully")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print("Withdrawal successful")
        else:
            print("Insufficient balance")

    def check_balance(self):
        print("Current Balance:", self.balance)


class ATM:
    FILE_NAME = "accounts.txt"

    def __init__(self):
        if not os.path.exists(self.FILE_NAME):
            open(self.FILE_NAME, "w").close()

    def save_account(self, account):
        with open(self.FILE_NAME, "a") as f:
            f.write(f"{account.acc_no},{account.name},{account.pin},{account.balance}\n")

    def get_account(self, acc_no, pin):
        with open(self.FILE_NAME, "r") as f:
            for line in f:
                data = line.strip().split(",")
                if data[0] == acc_no and data[2] == pin:
                    return Account(data[0], data[1], data[2], int(data[3]))
        return None

    def update_account(self, account):
        lines = []
        with open(self.FILE_NAME, "r") as f:
            lines = f.readlines()

        with open(self.FILE_NAME, "w") as f:
            for line in lines:
                data = line.strip().split(",")
                if data[0] == account.acc_no:
                    f.write(f"{account.acc_no},{account.name},{account.pin},{account.balance}\n")
                else:
                    f.write(line)


def main():
    atm = ATM()

    while True:
        print("\n--- ATM MENU ---")
        print("1. Create Account")
        print("2. Login")
        print("3. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            acc_no = input("Account Number: ")
            name = input("Name: ")
            pin = input("PIN: ")
            balance = int(input("Initial Balance: "))

            account = Account(acc_no, name, pin, balance)
            atm.save_account(account)
            print("Account created successfully")

        elif choice == "2":
            acc_no = input("Account Number: ")
            pin = input("PIN: ")

            account = atm.get_account(acc_no, pin)

            if not account:
                print("Invalid account or PIN")
                continue

            while True:
                print("\n1. Deposit")
                print("2. Withdraw")
                print("3. Check Balance")
                print("4. Logout")

                ch = input("Enter choice: ")

                if ch == "1":
                    amount = int(input("Amount: "))
                    account.deposit(amount)
                    atm.update_account(account)

                elif ch == "2":
                    amount = int(input("Amount: "))
                    account.withdraw(amount)
                    atm.update_account(account)

                elif ch == "3":
                    account.check_balance()

                elif ch == "4":
                    print("Logged out")
                    break

        elif choice == "3":
            print("Thank you for using ATM")
            break

        else:
            print("Invalid choice")


if __name__ == "__main__":
    main()
```
